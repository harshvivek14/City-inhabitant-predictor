```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('matched_city.csv')
city_inhabitant = df.values.tolist()
print(df.head)
```

```python
# function for generation of list of k last letters

def last_k_letters(city_inhabitant, k):
    n = len(city_inhabitant)
    fin = []
    end_char = []
    for i in range(n):
        j = 0
        while(j < len(city_inhabitant[i][0]) and j < len(city_inhabitant[i][1])):

            if city_inhabitant[i][0][j] != city_inhabitant[i][1][j]:
                fin.append([city_inhabitant[i][0],city_inhabitant[i][1],city_inhabitant[i][0][j-k+1:],city_inhabitant[i][1][j-k+1:]])
                end_char.append([city_inhabitant[i][0][j-k+1:],city_inhabitant[i][1][j-k+1:]])
                break
            j+=1
            if j == len(city_inhabitant[i][0]) and j<len((city_inhabitant[i][1])):
                fin.append([city_inhabitant[i][0],city_inhabitant[i][1],city_inhabitant[i][0][-k:],city_inhabitant[i][1][j-k:]])
                end_char.append([city_inhabitant[i][0][-k:],city_inhabitant[i][1][j-k:]])
                break
        
    return end_char
```

```python
# function for generation frequency dict

def freq_dic_of_letters(end_char_list):
    d = {}
    for i in end_char_list:
        if i[0] in d:
            d[i[0]] += 1
        else:
            d[i[0]] = 1

    sorted_d = sorted(d.items(), key=lambda x:-x[1])

    
    freq_dic = {}
    for i in end_char_list:
        city_name = i[0]
        inhabitant_name = i[1]
        if city_name in freq_dic:
            if inhabitant_name in freq_dic[city_name]:
                freq_dic[city_name][inhabitant_name] += 1
            else:
                freq_dic[city_name][inhabitant_name] = 1
        else:
            freq_dic[city_name] = {inhabitant_name: 1}
            
    sorted_freq_dic = sorted(freq_dic.items(), key=lambda x: -sum(x[1].values()))

    return sorted_d, sorted_freq_dic
```

```python
# Generating the dicts to define rules

two_end_char_list = last_k_letters(city_inhabitant, 2)
two_freq_dict = freq_dic_of_letters(two_end_char_list)

three_end_char_list = last_k_letters(city_inhabitant, 3)
three_freq_dict = freq_dic_of_letters(three_end_char_list)
```
